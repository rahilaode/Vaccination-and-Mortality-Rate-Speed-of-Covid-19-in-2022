```python
import pandas as pd
```

```python
penduduk = pd.read_excel('Jumlah Penduduk Menurut Provinsi di Indonesia.xlsx')
```

```python
penduduk.columns
```

```python
penduduk
```

```python
penduduk = penduduk.dropna()
```

```python
penduduk = penduduk[:-1]
```

```python
penduduk['2021'] = penduduk['Unnamed: 3']*1000
```

```python
penduduk
```

```python
covid = pd.read_csv('covid_19_indonesia_time_series_all (clean).csv')
```

```python
covid_new = pd.merge(covid, penduduk, on=['Province'])
covid_new
```

```python
covid_new['Mortality Rate'] = covid_new['Total Deaths'] / covid_new['2021'] * 1000
covid_new
```

```python
covid_new['Date'] = pd.to_datetime(covid_new['Date'])
```

```python
covid_new['Date'].head()
```

```python
covid_new[covid_new['Province'] == 'Sulawesi Tenggara'].Date.min()
```

```python
list_date_awal = []
province = []
for i in covid_new['Province'].unique():
    province.append(i)
    list_date_awal.append(covid_new[covid_new['Province'] == i].Date.min())

data = {'Province' : province, 'Start Date' : list_date_awal}
df_date_awal = pd.DataFrame(data)
df_date_awal
```

```python
covid_new = pd.merge(covid_new, df_date_awal, on=['Province'])
covid_new.head()
```

```python
covid_new.info()
```

```python
covid_new['Total Days'] = (covid_new['Date'] - covid_new['Start Date']).dt.days
covid_new.head()
```

```python
covid_new['Mortality Rate Speed'] = covid_new['Mortality Rate'] / covid_new['Total Days']
covid_new.head()
```

```python
covid_new = covid_new[covid_new['Total Days'] != 0]
```

```python
province = []
mortality_rate_speed_mean = []

for i in covid_new['Province'].unique():
    province.append(i)
    mortality_rate_speed_mean.append(covid_new['Mortality Rate Speed'][covid_new['Province'] == i].mean())
    
data = {'Province' : province, 'Mortality Rate Speed Mean' : mortality_rate_speed_mean}
mr_speed_mean = pd.DataFrame(data)
mr_speed_mean = mr_speed_mean.sort_values(['Mortality Rate Speed Mean'], ascending=False)
mr_speed_mean.head()
```

```python
mr_speed_mean.to_excel("Mortality Rate Speed (Rata-Rata).xlsx") 
```
