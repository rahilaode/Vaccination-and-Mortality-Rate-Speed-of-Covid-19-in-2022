### 3.1.2. Ambil kolom yang diperlukan

```python
df_covid = df_covid[['Date', 'Province', 'Population', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths', 
               'Total Recovered', 'Total Active Cases', 'Case Fatality Rate', 'Case Recovered Rate']]
df_covid.head()
```

```python
df_covid['Date']= pd.to_datetime(df_covid['Date'])
```

```python
df_covid.head()
```

```python
df_covid.info()
```

### 3.1.4. Rubah Tipe data Kolom Case Fatality Rate dan Case Recovered Rate Menjadi Float

```python
df_covid['Case Fatality Rate'] = list(map(lambda x: x[:-1], df_covid['Case Fatality Rate'].values))
df_covid['Case Recovered Rate'] = list(map(lambda x: x[:-1], df_covid['Case Recovered Rate'].values))
```

```python
df_covid['Case Fatality Rate'] = [float(x) for x in df_covid['Case Fatality Rate'].values]
df_covid['Case Recovered Rate'] = [float(x) for x in df_covid['Case Recovered Rate'].values]
df_covid.head()
```

```python
df_covid.info()
```

### 3.1.5. Rubah Nama Kolom Case Fatality Rate dan Case Recovered Rate

```python
df_covid.rename(columns = {'Case Fatality Rate':'Case Fatality Rate (%)', 'Case Recovered Rate':
                                      'Case Recovered Rate (%)'}, inplace = True)
df_covid.head()
```

## 3.7. Penambahan Kolom Mortality Rate (Per 1000 Populasi)

```python
df['Mortality Rate (2019)'] = df['Total Deaths'] / df['Population 2019'] * 1000
df['Mortality Rate (2020)'] = df['Total Deaths'] / df['Population 2020'] * 1000
df['Mortality Rate (2021)'] = df['Total Deaths'] / df['Population 2021'] * 1000
df.head()
```

## 3.8. Penambahan Kolom Vaccination Rate (Per 1000 Populasi)

```python
df['Vaccination Rate'] = df['total_vaksin_1'] / df['Population 2021'] * 1000
df.head()
```

## 3.9. Cek 5 Provinsi dengan Rata-Rata Mortality Rate Speed Tertinggi

### 3.9.1. Penambahan Kolom Start Date

```python
list_date_awal = []
province = []
for i in df['Province'].unique():
    province.append(i)
    list_date_awal.append(df[df['Province'] == i].Date.min())

data = {'Province' : province, 'Start Date' : list_date_awal}
df_date_awal = pd.DataFrame(data)
df_date_awal.head()
```

### 3.9.2. Menambahkan Kolom Start Date Pada DataFrame

```python
df = pd.merge(df, df_date_awal, on=['Province'])
df.head()
```

### 3.9.3. Penambahan Kolom Total Days Pada Dataframe

```python
df['Total Days'] = (df['Date'] - df['Start Date']).dt.days
df.head()
```

### 3.9.4. Penambahan Kolom Mortality Rate Speed

```python
df['Mortality Rate Speed'] = df['Mortality Rate (2021)'] / df['Total Days']
df.head()
```

### 3.9.5. Menghapus Baris Infinite  Pada Kolom Mortality Rate Speed (2021)

```python
df = df[df['Total Days'] != 0]
df.head()
```

### 3.9.6. Cek 5 Provinsi dengan Nilai Rata-Rata Mortality Rate Speed Tertinggi

```python
province = []
mortality_rate_speed_mean = []

for i in df['Province'].unique():
    province.append(i)
    mortality_rate_speed_mean.append(df['Mortality Rate Speed'][df['Province'] == i].mean())
    
data = {'Province' : province, 'Average Mortality Rate Speed' : mortality_rate_speed_mean}
mr_speed_mean = pd.DataFrame(data)
mr_speed_mean = mr_speed_mean.sort_values(['Average Mortality Rate Speed'], ascending=False)
mr_speed_mean.head()
```

# 4. Uji Hipotesis (Metode Causal Impact)

## 4.1. DKI Jakarta

```python
df_jkt = df_jkt.dropna(subset=['Mortality Rate Speed'])
df_jkt
```

```python
df_jkt.describe()
```

```python
#Tahun 2022
df_jkt_2022 = df_jkt[df_jkt['Date'] > '2021-12-31']
```

```python
df_jkt_2022.head()
```

```python
#Sebelum tahun 2022
df_jkt_before2022 = df_jkt[df_jkt['Date'] < df_jkt_2022.Date.min()]
```

```python
df_jkt_before2022.head()
```

```python
# Set pre-period
pre_period = [str(df_jkt_before2022.Date.min())[:10], str(df_jkt_before2022.Date.max())[:10]]

# Set post-period
post_period = [str(df_jkt_2022.Date.min())[:10], str(df_jkt_2022.Date.max())[:10]]

# Print out the values
print(f'The pre-period is {pre_period}')
print(f'The post-period is {post_period}')
```

```python
# Visualize data using seaborn
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=df_jkt.Date, y=df_jkt['Mortality Rate Speed'])
plt.axvline(x= df_jkt['Date'][df_jkt['Date'] == df_jkt_2022.Date.min()], color='red')
```

```python
df_jkt_new = df_jkt[['Date', 'Mortality Rate Speed']]
df_jkt_new = df_jkt_new.set_index('Date')
df_jkt_new.head()
```

```python
from causalimpact import CausalImpact
# Causal impact model
impact_jkt = CausalImpact(data=df_jkt_new, pre_period=pre_period, post_period=post_period)

# Visualization
impact_jkt.plot()
```

```python
# Causal impact summary
print(impact_jkt.summary())
```

```python
print("P Value DKI Jakarta = {}".format(impact_jkt.p_value))
pvalue_jkt = impact_jkt.p_value
```
