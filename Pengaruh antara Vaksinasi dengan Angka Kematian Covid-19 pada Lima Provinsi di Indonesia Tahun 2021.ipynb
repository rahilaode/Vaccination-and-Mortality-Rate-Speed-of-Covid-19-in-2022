## Uji Hipotesis

### Metode Causal Impact

#### Set Pre and Post Periods

```python
# Sumber sultra.bps.go.id
total_penduduk_sultra = 2755589
```

```python
df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)]
```

Proses vaksinasi mencapai 50% dari jumlah penduduk di sulawesi tenggara terhitung sejak tanggal 28 Desember 2021

```python
# Set pre-period
pre_period = [str(df_sultra.Date.min())[:10], '2021-12-27']

# Set post-period
post_period = ['2021-12-28', str(df_sultra.Date.max())[:10]]

# Print out the values
print(f'The pre-period is {pre_period}')
print(f'The post-period is {post_period}')
```

```python
df_sultra.reset_index(inplace=True)
df_sultra = df_sultra.drop('index', axis=1)
df_sultra.head()
```

```python
df_sultra[df_sultra.Date == '2021-12-28'].index[0]
```

```python
# Visualize data using seaborn
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=df_sultra.Date, y=df_sultra.New_Deaths)
plt.axvline(x= df_sultra.Date[df_sultra[df_sultra.Date == '2021-12-28'].index[0]], color='red')
```

```python
df_sultra_deaths = df_sultra[['Date','New_Deaths']]
df_sultra_deaths = df_sultra_deaths.set_index('Date')
df_sultra_deaths
```

#### Causal Impact Analysis

```python
from causalimpact import CausalImpact
# Causal impact model
impact = CausalImpact(data=df_sultra_deaths, pre_period=pre_period, post_period=post_period)

# Visualization
impact.plot()
```

```python
# Causal impact summary
print(impact.summary())
```

### Metode Anova

```python
df_sultra.head()
```

```python
df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)]
#len(df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)])
```

```python
treatment = df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)]
treatment.head()
```

```python
treatment.tail()
```

```python
control = df_sultra[657-len(treatment):657]
control.head()
```

```python
print(control.shape)
print(treatment.shape)
```

```python
from scipy.stats import f_oneway
```

```python
statistic, p_value = f_oneway(control['New_Deaths'], treatment['New_Deaths'])
```

```python
print(p_value)

if p_value >= 0.05:
    print('Failed to reject H0')
else:
    print('Reject H0')
```

### Metode Z-Test

```python
from statsmodels.stats.weightstats import ztest
```

```python
tstat, p_value = ztest(control['New_Deaths'], treatment['New_Deaths'])
```

```python
print('P-Value : ', p_value)
if p_value >= 0.05:
    print('Failed to reject H0')
else :
    print('Reject H0')
```
