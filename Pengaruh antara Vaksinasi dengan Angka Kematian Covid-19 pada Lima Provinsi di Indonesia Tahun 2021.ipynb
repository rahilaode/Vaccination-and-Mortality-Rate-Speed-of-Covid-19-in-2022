### Rubah Tipe data Kolom Case Fatality Rate dan Case Recovered Rate Menjadi Float

```python
df_covid['Case Fatality Rate'] = list(map(lambda x: x[:-1], df_covid['Case Fatality Rate'].values))
df_covid['Case Recovered Rate'] = list(map(lambda x: x[:-1], df_covid['Case Recovered Rate'].values))
```

```python
df_covid['Case Fatality Rate'] = [float(x) for x in df_covid['Case Fatality Rate'].values]
df_covid['Case Recovered Rate'] = [float(x) for x in df_covid['Case Recovered Rate'].values]
df_covid.head()
```

## Cek 5 Provinsi dengan Cause-Spesific Mortality Rate Tertinggi

```python
province = []
mortality_rate = []
total_deaths = []
population = []
start_date = []
end_date = []
for i in df['Province'].unique():
    start_date.append(df[df['Province']== i ]['Date'].min())
    end_date.append(df[df['Province']== i ]['Date'].max())
    province.append(i)
    total_deaths.append(df[df['Province']== i ]['Total Deaths'].max())
    population.append(df[df['Province']== i ]['Population'].median())
    mortality_rate.append((df[df['Province']== i ]['Total Deaths'].max()) * 1000 / 
                         (df[df['Province']== i ]['Population'].median()))
```

```python
data = {'Start Date' : start_date, 'End Date' : end_date , 'Province' : province, 'Total Deaths' : total_deaths, 
        'Population' : population,'Mortality Rate' : mortality_rate}
mr_prov = pd.DataFrame(data)
mr_prov = mr_prov.sort_values(['Mortality Rate'], ascending=False)
mr_prov.head()
```

### Metode Anova

```python
df_sultra.head()
```

```python
df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)]
#len(df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)])
```

```python
treatment = df_sultra[df_sultra.Vaccinated_Cumulative >= (total_penduduk_sultra/2)]
treatment.head()
```

```python
treatment.tail()
```

```python
control = df_sultra[657-len(treatment):657]
control.head()
```

```python
print(control.shape)
print(treatment.shape)
```

```python
from scipy.stats import f_oneway
```

```python
statistic, p_value = f_oneway(control['New_Deaths'], treatment['New_Deaths'])
```

```python
print(p_value)

if p_value >= 0.05:
    print('Failed to reject H0')
else:
    print('Reject H0')
```

### Metode Z-Test

```python
from statsmodels.stats.weightstats import ztest
```

```python
tstat, p_value = ztest(control['New_Deaths'], treatment['New_Deaths'])
```

```python
print('P-Value : ', p_value)
if p_value >= 0.05:
    print('Failed to reject H0')
else :
    print('Reject H0')
```
