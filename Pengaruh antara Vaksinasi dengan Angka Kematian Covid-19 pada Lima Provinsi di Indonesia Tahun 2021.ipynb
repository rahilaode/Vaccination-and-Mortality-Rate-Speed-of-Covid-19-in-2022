## 3.1Data Covid

```python
df_covid.head()
```

### 3.1.1. Ambil kolom yang diperlukan

```python
df_covid = df_covid[['Date', 'Province', 'Population', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths', 
               'Total Recovered', 'Total Active Cases', 'Case Fatality Rate', 'Case Recovered Rate']]
df_covid.head()
```

### 3.1.2. Rubah Kolom Date Menjadi Datetime

```python
df_covid['Date']= pd.to_datetime(df_covid['Date'])
```

```python
df_covid.head()
```

```python
df_covid.info()
```

### 3.1.3. Rubah Tipe data Kolom Case Fatality Rate dan Case Recovered Rate Menjadi Float

```python
df_covid['Case Fatality Rate'] = list(map(lambda x: x[:-1], df_covid['Case Fatality Rate'].values))
df_covid['Case Recovered Rate'] = list(map(lambda x: x[:-1], df_covid['Case Recovered Rate'].values))
```

```python
df_covid['Case Fatality Rate'] = [float(x) for x in df_covid['Case Fatality Rate'].values]
df_covid['Case Recovered Rate'] = [float(x) for x in df_covid['Case Recovered Rate'].values]
df_covid.head()
```

```python
df_covid.info()
```

### 3.1.4. Rubah Nama Kolom Case Fatality Rate dan Case Recovered Rate

```python
df_covid.rename(columns = {'Case Fatality Rate':'Case Fatality Rate (%)', 'Case Recovered Rate':
                                      'Case Recovered Rate (%)'}, inplace = True)
df_covid.head()
```

```python
Date = df_vaksin1["Tanggal"].str.split(" ", n = 1, expand = True)[0]
df_vaksin1["Date"] = Date
df_vaksin1
```

## Penggabungan DataFrame

```python
df = pd.merge(df_covid, df_vaksin1, on=['Date', 'Province'], how='outer')
df = pd.merge(df, df_vaksin2, on=['Date', 'Province'], how='outer')
df = pd.merge(df, df_vaksin3, on=['Date', 'Province'], how='outer')
df = pd.merge(df, df_jml_penduduk, on=['Province'], how='outer')
df.head()
```

## Penambahan Kolom Mortality Rate (Per 1000 Populasi)

```python
df['Mortality Rate (2019)'] = df['Total Deaths'] / df['Population 2019'] * 1000
df['Mortality Rate (2020)'] = df['Total Deaths'] / df['Population 2020'] * 1000
df['Mortality Rate (2021)'] = df['Total Deaths'] / df['Population 2021'] * 1000
df.head()
```

## Cek 5 Provinsi dengan Rata-Rata Mortality Rate Speed Tertinggi

### Penambahan Kolom Start Date

```python
list_date_awal = []
province = []
for i in df['Province'].unique():
    province.append(i)
    list_date_awal.append(df[df['Province'] == i].Date.min())

data = {'Province' : province, 'Start Date' : list_date_awal}
df_date_awal = pd.DataFrame(data)
df_date_awal.head()
```

### Menambahkan Kolom Start Date Pada DataFrame

```python
df = pd.merge(df, df_date_awal, on=['Province'])
df.head()
```

### Penambahan Kolom Total Days Pada Dataframe

```python
df['Total Days'] = (df['Date'] - df['Start Date']).dt.days
df.head()
```

### Penambahan Kolom Mortality Rate Speed

```python
df['Mortality Rate Speed (2021)'] = df['Mortality Rate (2021)'] / df['Total Days']
df.head()
```

### Menghapus Baris Infinite  Pada Kolom Mortality Rate Speed (2021)

```python
df = df[df['Total Days'] != 0]
df.head()
```

### Cek 5 Provinsi dengan Nilai Rata-Rata Mortality Rate Speed (2021) Tertinggi

```python
province = []
mortality_rate_speed_mean = []

for i in df['Province'].unique():
    province.append(i)
    mortality_rate_speed_mean.append(df['Mortality Rate Speed (2021)'][df['Province'] == i].mean())
    
data = {'Province' : province, 'Mortality Rate Speed (2021) Mean' : mortality_rate_speed_mean}
mr_speed_mean = pd.DataFrame(data)
mr_speed_mean = mr_speed_mean.sort_values(['Mortality Rate Speed (2021) Mean'], ascending=False)
mr_speed_mean.head()
```
